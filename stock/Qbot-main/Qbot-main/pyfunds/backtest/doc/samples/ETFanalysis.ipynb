```python
path = "../../tests/etf.csv"
read = xa.record(path, skiprows=1)
# 读取E大的发车记录，该记录采取且慢的发车记录，但由于早期有些仓位是按点数不定时买入的，可能买入时点与E大和且慢记录稍有出入
```
